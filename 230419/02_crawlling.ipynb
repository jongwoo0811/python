```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
# 답변에 <Response [200]> 정상적으로 호출했다는 뜻
# 실제 사이트에서 html 정보를 받아와서 변수에 저장
# requests.get("https://finance.naver.com/marketindex/exchangeList.naver")
# .text로 데이터 가져오기
data = requests.get("https://finance.naver.com/marketindex/exchangeList.naver").text
```

```python
# BeautifilSoup을 이용하여 데이터를 파싱
soup =bs(data, 'html.parser')
```

```python
#print(soup)
```

```python
# tbody값이 하나만 있는 웹사이트라서 find로 tbody찾기
# find는 html안에서 해당 태그의 첫번째 정보를 출력
soup.find('tbody')
```

```python
# find_all() 해당 태그의 전체 정보를 리스트 형태로 출력
data_list = soup.find_all('td')#\n 은 줄바꿈 \t는 tab
```

```python
# soup.find_all('td')[0].get_text().repalce('\n','').strip('\t') 에러남;
# data_list를 이용하여 2차원 배열로 새로운 리스트를 생성
_list =[]
for i in data_list:
    x= i.get_text()
    x=x.replace("\n","")
    x=x.strip("\t")
    _list.append(x)
_list
```

```python
result =[]
for i in range(len(_list)//7):
    result.append(_list[i*7:(i+1)*7])
result
```
