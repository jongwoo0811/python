```python
import pandas as pd
```

```python
df = pd.read_csv("http://bit.ly/ds-house-price")
df
```

```python
# 컬럼의 이름을 단일로만 변경하는 방법
df= df.rename(columns={'분양가격(㎡)': '분양가격'})
df.info()
```

1. 분양가격 컬럼의 데이터의 형태를 int의 형태로변경
2. 규모구분 컬럼의 "전용면적" 단어를 삭제
3. 지역, 규모구분 별 분양가격의 평균가

```python
df.loc[df['분양가격']== '  ']
```

```python
# 시리즈 형태가 그냥 strip을 사용할 수 없어 문자열로 변경 후 스트립사용
df['분양가격'] = df['분양가격'].str.strip()
df['분양가격'] = df['분양가격'].str.replace(",", "")# 컴마 지우기
df['분양가격'] = df['분양가격'].str.replace('-', '')# - 지우기
df.loc[df['분양가격'] == "", '분양가격'] = 0
df['분양가격'].fillna(0, inplace=True)
```

```python
df['분양가격'] = df['분양가격'].astype(int)
```

```python
# 규모구분에서 '전용면적'ㅇ을 삭제
df['규모구분'] = df['규모구분'].str.replace('전용면적 ', '')
```

```python
df.groupby(['지역명', '규모구분']).mean()[['분양가격']]
```
