### 문제
1. test_csv폴더 안에 있는 오프라인 .csv를 로드
2. 컬럼의 이름을 ['기준년월','업종','성별','연령대','총이용건수','총이용급액']으로 변경
3. 기준년월을 시계열 데이터로 변경
4. index를 기준년월로 변경
5. 온라인.csv파일도 위의 작업을 반복한다.
6. 온라인에 업종 컬럼을 추가하여 '온라인'으로 값을 지정
7. 두개의 데이터프레임을 유니언 결합
8. 업종별로 총이용금액의 합계를 그래프로 표시

```python
import pandas as pd
```

```python
df = pd.read_csv("../test_csv/오프라인.csv",encoding='cp949')
df.head(1)
```

```python
df.columns= ['기준년월','업종','성별','연령대','총이용건수','총이용금액']
```

```python
df['기준년월'] = pd.to_datetime(df['기준년월'], format = "%Y%m")
df.info()
```

```python
df.set_index('기준년월',inplace=True)
```

```python
df
```

```python
df1 = pd.read_csv("../test_csv/온라인.csv")
df1.head(1)
```

```python
df1['업종']= '온라인'
```

```python
df1.columns= ['기준년월','성별','연령대','총이용건수','총이용금액','업종']
df1
```

```python
df1['기준년월'] = pd.to_datetime(df1['기준년월'], format = "%Y%m")
df1.info()
```

```python
df1.set_index('기준년월', inplace=True)
```

```python
df1
```

```python
#유니언 결합
data = pd.concat([df,df1],axis=0)
```

```python
data.sort_index(inplace=True)
```

```python
pivot_data = pd.pivot_table(
    data,
    index =data.index,
    columns='업종',
    values='총이용금액',
    aggfunc='sum'
)
```

```python
plt.rc('font', family ='Malgun Gothic')
```

```python
pivot_data.plot()
```
