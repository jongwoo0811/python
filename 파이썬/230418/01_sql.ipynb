### sql 그룹화
1. Country, Item Type을 기준으로 그룹화
2. Unit Sold의 평균과, 합계 생성
3. Country를 기준으로 오름차순 정렬

```python
sql ="""
    select
    `Country`,
    `Item Type`,
    AVG(`Units Sold`)as AVG,
    SUM(`Units Sold`)as SUM
    from
    `Sales Records`
    group by
    `Country`,
    `Item Type`
    order by
    `Country`
    ASC 
"""
db.sql_query(sql)
#엑셀과 같은 평균 구하는 AVG
#ASC 어센딩
```

2. job_code SALESMANS이면 1, MANAGER면 2 나머지는 3으로 하여 새로운 파생변수에 대입

```python
emp['JOB_CODE'] = 3
for i in emp.index:
    if emp.loc[i,'JOB'] == 'SALESMAN' :
        emp.loc[i, 'JOB_CODE'] =1
    elif emp.loc[i,'JOB'] == 'MANAGER':
        emp.loc[i, 'JOB_CODE'] =2

emp

# apply lambda 사용
# emp['JOB'].apply(lambda x: 1 if (X=='SALESMAN') else(2 if (x=='MANAGER)else 3))
```

```python
#sql 문 사용
# case 를 끝내려면 end 사용 후에 컬럼의 이름을 저장하려면 end as 컬럼이름
sql = """
    select
    `EMPNO`,
    `ENAME`,
    `JOB`,
    CASE
        when `JOB` = 'SALESMAN'
        then 1
        when `JOB` = 'MANAGER'
        then 2
        else 3
    end as `JOB_CODE`
    from emp
"""
db.sql_query(sql)
```

```python
sql = """
    select
    `EMPNO`,
    `ENAME`,
    `JOB`,
    if(`JOB`= 'SALESMAN', 1,
      if(`JOB` = 'MANAGER',2,3)
    )as `JOB_CODE`
    from
    emp
"""
db.sql_query(sql)
```

```python
sql = """
    select *
    from (select * from tran_1 union select * from tran_2) as tran
    left join
    (select * from tran_d_1 union select * from tran_d_2) as tran_d
    on tran.transaction_id = tran_d.transaction_id
"""
db.sql_query(sql)
```

### 문자열이 포함되어 있는 경우 출력

```python
# 문자열에 s가 포함되어 있는 것들만 출력'%%S%%'
# '%%S = s로끝나는
# S%%' = s로 시작하는
sql ="""
    select *
    from emp
    where ENAME like '%%S%%'
"""
db.sql_query(sql)
```

1. 판다스를 이용하여 emp.csv에서 ENAME이 S로 시작하는 데이터를 출력
2. S로 끝나는 데이터를 출력
3. S가 포함되어있는 데이터를 출력

```python
# apply(lambda) 사용
emp.loc[emp['ENAME'].apply(lambda x: x.startswith('S'))]
# apply(lambda) 사용
emp.loc[emp['ENAME'].apply(lambda x: x.endswith('S'))]
```

```python
# 간단 / 자동화 불가
emp.loc[emp['ENAME'].str.startswith('S')]
emp.loc[emp['ENAME'].str.endswith('S')]
emp.loc[emp['ENAME'].str.contains('S')]
```
